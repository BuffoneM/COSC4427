```python
# Codes of the loading "the_bible.txt" file goes here
import numpy as np

fileName = "The_Bible.txt"
file = open(fileName, "r")
```

# <span style="color:Green">P1: Count the number of characters.</span>

```python
#Codes of P1 goes here
def countChars(file):
    count = 0
    for line in file:
            for word in line.split():
                count += len(word)
    return count

with open(fileName, "r") as file:
    print("Number of characters:", countChars(file))
```

# <span style="color:Green">P2: Count the number of words (separated by whitespaces, tabs, and newlines characters).</span>

```python
#Codes of P2 goes here
def countWords(file):
    count = 0
    for line in file:
            for word in line.split():
                count += 1
    return count

with open(fileName, "r") as file:
    print("Number of words:", countWords(file))
```

# <span style="color:Green">P3: Count the number of unique words (separated by whitespaces, tabs, and newlines characters).</span>

```python
#Codes of P3 goes here
def countUniqueWords(file):
    words = []
    for line in file:
        for word in line.split():
            words.append(word)
    return len(np.unique(words))

with open(fileName, "r") as file:
    print("Number of unique words:", countUniqueWords(file))
```

# <span style="color:Green">P4: Count the number of lines.</span>

```python
#Codes of P4 goes here
def countLines(file):
    count = 0
    for line in file:
            count += 1
    return count

with open(fileName, "r") as file:
    print("Number of lines:", countLines(file))
```

# <span style="color:Green">P5: Find and print the longest word.</span>

```python
#Codes of P5 goes here
def longestWord(file):
    words = []
    for line in file:
        for word in line.split():
            words.append(word)
    longest = max(words, key=len)       
    print("Longest word:", longest, "-", len(longest), "characters long")
    
with open(fileName, "r") as file:
    longestWord(file)
```

# <span style="color:Green">P6: Find and print the 10 most common words.</span>

```python
#Codes of P6 goes here
def mostCommon(file):
    words = []
    for line in file:
        for word in line.split():
            words.append(word)
    
    word, count = np.unique(words, return_counts = True)
    result = sorted(zip(count, word))
    print(result[-10:])
    
with open(fileName, "r") as file:
    print("Most common words:")
    mostCommon(file)
```

# <span style="color:Green">P7: Find and print the 10 least common words.</span>

```python
#Codes of P7 goes here
def leastCommon(file):
    words = []
    for line in file:
        for word in line.split():
            words.append(word)
    
    word, count = np.unique(words, return_counts = True)
    result = sorted(zip(count, word))
    print(result[:10])
    
with open(fileName, "r") as file:
    print("Least common words:")
    leastCommon(file)
```

# <span style="color:Green">P8: Count the number of words that begin with each English letter.</span>

```python
#Codes of P8 goes here
def startWithLetter(file):
    count = 0
    for line in file:
            for word in line.split():
                if word[0].isalpha():
                    count += 1
    return count

with open(fileName, "r") as file:
    print("Words that start with letters:", startWithLetter(file))
```

# <span style="color:Green">P9: List unique words by each English letter.</span>

```python
#Codes of P9 goes here
def listByLetter(file):
    words = []
    for line in file:
        for word in line.split():
            words.append(word)
            
    uniqueWords = np.unique(words)
    for item in uniqueWords:
        if item[0].isalpha():
            print(item)

with open(fileName, "r") as file:
    print("Unique words that start with only English letters:")
    listByLetter(file)
```

# <span style="color:Green">P10: Count the number of digits. </span>

```python
#Codes of P10 goes here
def countDigits(file):
    count = 0
    for line in file:
            for word in line.split():
                for currChar in word:
                    if currChar.isdigit():
                        count += 1
    return count

with open(fileName, "r") as file:
    print("The number of digits as characters are:", countDigits(file))
```
